ADF 검정은 시계열 데이터가 안정적인 패턴을 가지고 있는지 확인하는 방법입니다. 여기서 시계열 데이터란 시간의 흐름에 따라 기록된 데이터로, 예를 들면 월별 매출, 일별 기온, 주가 같은 것입니다.  

시계열 분석에서는 데이터가 정상적(stationary) 인 상태인지가 중요합니다. 정상적이라는 것은 시간이 지나도 데이터의 전반적인 성격이 크게 바뀌지 않는다는 뜻입니다.  
예를 들어 평균 수준이 갑자기 계속 올라가거나 내려가지 않고, 변동 폭도 비슷하게 유지되는 상태를 말합니다.

반대로 비정상적인 데이터는 시간이 지남에 따라 추세가 생기거나, 계절에 따라 반복적으로 변하거나, 변동 폭이 점점 커지는 경우입니다. 이런 데이터는 바로 분석하면 결과가 부정확할 수 있습니다.

검정의 기준은 다음과 같습니다:

- 귀무 가설 (H0): 이 데이터는 안정적이지 않다. 즉, 비정상적이다.
- 대립 가설 (H1): 이 데이터는 안정적이다. 즉, 정상적이다.  

검정 결과에서 귀무가설을 기각할 수 있으면, 보통 **“이 데이터는 정상성을 가진다”** 고 판단합니다.

시계열 데이터가 정상성을 가지지 않는 경우, 데이터 변환, 차분 등의 처리를 통해 정상성을 확보하고 이후 분석을 진행할 수 있습니다.  

- 귀무가설은 기본적으로 일단 사실이라고 가정해 두는 주장
